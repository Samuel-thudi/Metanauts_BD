```python
import pandas as pd
```

```python
d = open('shakespeare.txt')
```

```python
def split(data):
    return([i for item in data for i in item.split()])
```

```python
words = split(d)
```

```python
x = []
x = set(words)
```

```python
len(x)
```

```python
data = pd.DataFrame(x)
data.columns = ['words']
data["words"] = data['words'].str.lower()
```

```python
y = []

for i in data['words']:
    y.append(words.count(i))
```

```python
len(y)
```

```python
data['count'] = y
```

```python
data = data.sort_values('count',ascending=False)
```

```python
area_dict = dict(zip(data['words'], data['count']))
```
